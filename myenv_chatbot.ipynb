```python
import pandas as pd
import nlpaug.augmenter.word as naw

# Load your dataset
data = corpus

# Initialize the augmentation method
aug = naw.SynonymAug(aug_src='wordnet')

# Define the number of augmented samples to generate for each original sample
num_augmented_samples = 3

# Augment the data
augmented_data = []

for _, row in data.iterrows():
    input_text = row['input_text']  # Replace 'input_text' with the actual column name containing your input text
    response = row['response']  # Replace 'response' with the actual column name containing your response

    augmented_samples = aug.augment(input_text, n=num_augmented_samples)

    for augmented_sample in augmented_samples:
        augmented_data.append({'input_text': augmented_sample, 'response': response})

# Convert the augmented data to a DataFrame
augmented_df = pd.DataFrame(augmented_data)

# Concatenate the original data with the augmented data
augmented_dataset = pd.concat([data, augmented_df], ignore_index=True)
```

```python

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
from sklearn.feature_extraction.text import CountVectorizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load the dataset


# Function to preprocess text
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenize text
    tokens = word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    # Join tokens back into text
    preprocessed_text = ' '.join(tokens)
    return preprocessed_text

# Apply preprocessing to question and answer columns
new_data['question_clean'] = new_data['question'].apply(preprocess_text)
new_data['answer_clean'] = new_data['answer'].apply(preprocess_text)

print(new_data)
```

```python
########### GLOVE + BERT for  Embedding  and vectorization to retreive information #####################
import numpy as np
from torch.nn.utils.rnn import pad_sequence
# Load GloVe embeddings (you may need to download them first)
# Replace 'glove.6B.100d.txt' with the appropriate GloVe file
embedding_file = '/content/drive/MyDrive/glove.6B.100d.txt'
embeddings_index = {}
with open(embedding_file, encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs

# Tokenize input texts and responses
input_tokenized_texts = [tokenizer.encode(text, add_special_tokens=True) for text in input_texts]
response_tokenized_texts = [tokenizer.encode(text, add_special_tokens=True) for text in responses]

# Convert tokenized texts to GloVe embeddings
input_word_vectors = []
for tokens in input_tokenized_texts:
    vectors = [embeddings_index[token] if token in embeddings_index else np.zeros(100) for token in tokens]
    input_word_vectors.append(vectors)

response_word_vectors = []
for tokens in response_tokenized_texts:
    vectors = [embeddings_index[token] if token in embeddings_index else np.zeros(100) for token in tokens]
    response_word_vectors.append(vectors)

# Convert to PyTorch tensors
input_word_vectors = [torch.tensor(vectors) for vectors in input_word_vectors]
response_word_vectors = [torch.tensor(vectors) for vectors in response_word_vectors]

# Pad sequences
input_word_vectors = pad_sequence(input_word_vectors, batch_first=True, padding_value=0)
response_word_vectors = pad_sequence(response_word_vectors, batch_first=True, padding_value=0)
```
